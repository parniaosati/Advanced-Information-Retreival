    در این قسمت تابع clean_data() را پیاده سازی کنید. عملکرد تابع به این صورت است که یک متن به عنوان ورودی گرفته و توکن‌های ولید آن را به صورت یک لیست که عملیات lemmatization, stemming و case folding اجرا شده است خروجی می‌دهد. متن ورودی شامل عنوان مقاله یا چکیده آنهاست.

```python
## 6 points
import nltk

def clean_data(text : str):
    """Preprocesses the text with tokenization, case folding, stemming and lemmatization, and punctuations

    Parameters
    ----------
    text : str
        The title or abstract of an article

    Returns
    -------
    list
        A list of tokens
    """

    # TODO: tokenize, case_folding, stem, lemmatize, punctuations
    stemmer = nltk.stem.PorterStemmer()
    lemmatizer = nltk.stem.WordNetLemmatizer()

    tokens = nltk.word_tokenize(text.lower())
    tokens = [stemmer.stem(token) for token in tokens]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    tokens = nltk.regexp_tokenize(' '.join(tokens), r'\w+')

    return tokens

clean_data("An abstract is a summary of the main article.") # return ["an", "abstract", "is", "a", "summary", "of", "the", "main", "article"]
```

```python
## 2 points
import collections

def filtered_tokens(tokens, filters):
    return list(filter(lambda word: word not in filters, tokens))

def find_stop_words(all_text : list[str], num_token=30, print_occurences=False):
    """Detects stop-words

     Parameters
    ----------
    all_text : list of all tokens
        (result of clean_data(text) for all the text)

    Returns
    -------
    Return Value is optional but must print the stop words and number of their occurence
    """

    # TODO: find num_token top most repetitive terms and report them
    occurences = collections.Counter(all_text)
    stop_words = dict(occurences.most_common(num_token))
    filtered_words = filtered_tokens(all_text, stop_words.keys())

    if print_occurences:
        print(stop_words)
    return filtered_words, list(stop_words.keys())

find_stop_words(clean_data("An abstract is a summary of the main article."), 2, True)
```

```python
## 1 points
# TODO: read data file and do the preprocessing on the data (first clean the data and then detect and delete stop words)
import pandas as pd

data = pd.read_csv('data.csv')

titles = {}
abstracts = {}
title_tokens = {}
abstract_tokens = {}
for i, row in data.iterrows():
    paper_id = row['paperId']
    
    titles[paper_id] = str(row['title'])
    if titles[paper_id] == 'nan':
        titles[paper_id] = ''
    abstracts[paper_id] = str(row['abstract'])
    if abstracts[paper_id] == 'nan':
        abstracts[paper_id] = ''

    title_tokens[paper_id] = clean_data(titles[paper_id])
    abstract_tokens[paper_id] = clean_data(abstracts[paper_id])
```

```python
import itertools

all_tokens = list(title_tokens.values()) + list(abstract_tokens.values())
_, stop_words = find_stop_words(list(itertools.chain(*all_tokens)))

corpus = {}
for paper_id in data['paperId']:
    corpus[paper_id] = {
        'title': filtered_tokens(title_tokens[paper_id], stop_words),
        'abstract': filtered_tokens(abstract_tokens[paper_id], stop_words),
    }
```

```python
## 12 points
from collections import defaultdict

def construct_positional_indexes(corpus : str):
    
    """
    Get processed data and insert words in that into a trie and construct postional_index and posting lists after wards.

    Parameters
    ----------
    corpus: str
        processed data 
    
    Return
    ----------
    docs: 
        list of docs with specificied id, title, abstract.
    """

    postional_index = defaultdict(lambda: defaultdict(lambda: {'title': [], 'abstract': []}))
    for paper_id, details in corpus.items():
        for label in ['title', 'abstract']:
            for index, token in enumerate(details[label]):
                postional_index[token][paper_id][label].append(index)

    return postional_index

docs = construct_positional_indexes(corpus)
```

```python
## 7 points

def add_documnet(document : tuple):
    """Adds a document to positional index

    Parameters
    ----------
    document : str
        Comma separated string containing id,title,abstarct in this exact order
    """

    # TODO: implement adding a document to the index
    paper_id, title, abstract = document
    
    titles[paper_id] = titles
    abstracts[paper_id] = abstract
    title_tokens[paper_id] = clean_data(title)
    abstract_tokens[paper_id] = clean_data(abstract)
    corpus[paper_id] = {
        'title': filtered_tokens(title_tokens[paper_id], stop_words),
        'abstract': filtered_tokens(abstract_tokens[paper_id], stop_words),
    }

    for label in ['title', 'abstract']:
        for index, token in enumerate(corpus[paper_id][label]):
            docs[token][paper_id][label].append(index)

new_document = ("1eae26fe1ca566f17468080c3aecab1c3f9efb66", 
                "A Deep Learning Framework for Viable Tumor Burden Estimation", 
                "Liver masses have become a common clinical challenge since they require to be defined and accurately categorized as neoplastic or nonneoplastic lesions. Hepatocellular carcinoma (HCC), the most common histologic type of primary liver malignancy, is a global health concern being the fifth most common cancer and the second cause of cancer mortality worldwide. Accurate diagnosis, which in some circumstances requires histopathology results, is necessary for appropriate management. Also, some tumor characteristics help in predicting tumor behavior and patient response to therapy. In this paper, we propose a deep learning framework for the segmentation of whole and viable tumor areas of liver cancer from whole-slide images (WSIs). To this end, we use Fast Segmentation Convolutional Neural Network (Fast-SCNN) as our network. We use the dataset from PAIP 2019 challenge. After data-augmentation on the training subset, we train the network with a multi-term loss function and SWA technique. Our model achieves 0.80 for the median of the Jaccard Index for the task of Viable Tumor Segmentation and 0.77 for the median of Weighted Absolute Accuracy for the task of Viable Tumor Burden Estimation on the whole-slide images of the test subset.")
add_documnet(new_document)
```

```python
## 7 points

def remove_document(document_id : str):
    """removes a document from positional index

    Parameters
    ----------
    document_id : str
        Id of the document
    """

    # TODO: implement removing a document from the index
    del titles[document_id]
    del abstracts[document_id]
    del title_tokens[document_id]
    del abstract_tokens[document_id]

    for label in ['title', 'abstract']:
        for token in corpus[document_id][label]:
            if document_id in docs[token]:
                docs[token].pop(document_id)
    del corpus[document_id]
  
remove_document("1eae26fe1ca566f17468080c3aecab1c3f9efb66")
```

```python
import math

def gamma_encode(array):
    encoded = ''
    for num in array:
        binary = bin(num)[2:]
        zeros = '0' * len(binary)
        encoded += f'{zeros}1{binary}' if num else '1'
    return encoded

def gamma_decode(encoded):
    decoded = []
    while encoded:
        index = encoded.find('1')
        binary = encoded[index + 1:2 * index + 1]
        encoded = encoded[2 * index + 1:]
        decoded.append(int(binary, 2) if binary else 0)
    return decoded
```

```python
def varbyte_encode(array):
    encoded = bytearray()
    for num in array:
        while num:
            byte = num & 0x7F
            num >>= 7
            encoded.append(byte | 0x80 if num else byte)
    return encoded.decode('iso-8859-1')

def varbyte_decode(encoded):
    decoded = []
    num, shift = 0, 0
    encoded = encoded.encode('iso-8859-1')
    for byte in encoded:
        num |= (byte & 0x7F) << shift
        if byte & 0x80:
            shift += 7
        else:
            decoded.append(num)
            num, shift = 0, 0
    return decoded
```

```python
def get_encoded(positional_index, encoder):
    encoded = defaultdict(lambda: defaultdict(lambda: {'title': [], 'abstract': []}))
    for token, details in positional_index.items():
        for paper_id, data in details.items():
            for label in ['title', 'abstract']:
                encoded[token][paper_id][label] = encoder(data[label])
    return encoded

def get_decoded(positional_index, decoder):
    decoded = defaultdict(lambda: defaultdict(lambda: {'title': str, 'abstract': str}))
    for token, details in positional_index.items():
        for paper_id, data in details.items():
            for label in ['title', 'abstract']:
                decoded[token][paper_id][label] = decoder(data[label])
    return decoded
```

```python
## 13 points
import os
import json

def store_index(path: str, compression_type: str):
    """Stores the index in a file

    Parameters
    ----------
    path : str
        Path to store the file

    compression_type : str
        Could be one of the followings:
        - no-compression
        - gamma-code
        - variable-byte

    Returns
    int
        The size of the stored file
    """

    # TODO: compress and store positional_index
    if compression_type == 'gamma-code':
        with open(path, 'w') as file:
            json.dump(get_encoded(docs, gamma_encode), file)
    elif compression_type == 'variable-byte':
        with open(path, 'w') as file:
            json.dump(get_encoded(docs, varbyte_encode), file)
    else:
        with open(path, 'w') as file:
            json.dump(docs, file)

    size_of_file = os.path.getsize(path)
    return size_of_file

store_index("index.json", "no-compression")
```

```python
## 7 points

def load_index(path: str, compression_type: str):
    """Loads the index from a file

    Parameters
    ----------
    path : str
        Path of the file to load from

    compression_type : str
        Could be one of the followings:
        - no-compression
        - gamma-code
        - variable-byte
    """

    # TODO: load and decompress positional_index
    if compression_type == 'gamma-code':
        with open(path, 'r') as file:
            positional_index = get_decoded(json.load(file), gamma_decode)
    elif compression_type == 'variable-byte':
        with open(path, 'r') as file:
            positional_index = get_decoded(json.load(file), varbyte_decode)
    else:
        with open(path, 'r') as file:
            positional_index = json.load(file)

    return positional_index

docs = load_index("index.json", "no-compression")
```

```python
## 3 points

from nltk.util import ngrams
from typing import List, Set, Dict

def create_bigram_index(texts: List[str]) -> Dict[str, Set[str]]:
    """
    Creates a bigram index for the spell correction

    Parameters
    ----------
    texts: List[str]
        The titles and abstracts of articles

    Returns
    -------
    dict
        A dictionary of bigrams and their occurence
    """
    bigram: Dict[str, Set[str]] = {}

    # TODO: Create the bigram index here
    for token in itertools.chain(*[sentence.lower().split() for sentence in texts]):
        token = ''.join(filter(str.isalnum, token))
        for ngram in set(ngrams(token, 2)):
            string = ''.join(ngram)
            if string not in bigram:
                bigram[string] = set()
            bigram[string].add(token)

    return bigram

bigram_index = create_bigram_index(list(titles.values()) + list(abstracts.values()))
```

```python
## 9 points
from nltk.metrics import jaccard_distance, edit_distance

def correct_text(text: str, similar_words_limit: int=20) -> str:
    """
    Correct the give query text, if it is misspelled

    Paramters
    ---------
    text: str
        The query text
    
    Returns
    str
        The corrected form of the given text
    """
    correct_words = []

    # TODO: Correct the text here
    for word in text.lower().split():
        word_bigrams = set(ngrams(word, 2))

        similarities = set()
        for bigram in word_bigrams:
            string = ''.join(bigram)
            if string in bigram_index:
                for token in bigram_index[string]:
                    similarity = 1 - jaccard_distance(word_bigrams, set(ngrams(token, 2)))
                    similarities.add((similarity, token))

        similar_words = sorted(similarities, reverse=True)[:similar_words_limit]
        if len(similar_words) == 0:
            return ''
        correct_word = min(similar_words, key=lambda token: edit_distance(word, token[1]))[1]

        correct_words.append(correct_word)

    return ' '.join(correct_words)

print(correct_text('Transtation Model Baxed on Dep Lerning'))
```

```python
def preprocess_query(query):
    tokens = clean_data(correct_text(query))
    return filtered_tokens(tokens, stop_words)

def get_logarithmic_tf(tf):
    return 1 + math.log10(tf) if tf else 0

def get_idf(token):
    df = len(docs[token])
    return math.log10(len(corpus) / (df + 1))

def get_snippet(content, query):
    query_tokens = preprocess_query(query)
    content_tokens = preprocess_query(content)

    for start_index in range(len(content_tokens) - len(query_tokens) + 1):
        end_index = start_index + len(query_tokens)
        if content_tokens[start_index:end_index] == query_tokens:
            return '... ' + ' '.join(content.split()[start_index:end_index]) + ' ...'
    return ' '.join(content.split()[:20]) + ' ...' if len(content) > 100 else content

def get_doc_scores(query_count, method, where):
    doc_scores = defaultdict(int)
    norm_factors = defaultdict(int)
    average_doc_length = sum(len(details[where]) for details in corpus.values())

    for token, count in query_count.items():
        if token not in docs:
            continue

        for paper_id, details in docs[token].items():
            if not details[where]:
                continue

            if method == 'ltn-lnn' or method == 'ltc-lnc':
                doc_coeff = get_idf(token) * get_logarithmic_tf(len(details[where]))
                doc_scores[paper_id] += get_logarithmic_tf(count) * doc_coeff

                if method == 'ltc-lnc':
                    norm_factors[paper_id] += doc_coeff ** 2

            if method == 'okapi25':
                k1, b = 1.5, .75
                tf_token = len(details[where])
                doc_scores[paper_id] += (k1 + 1) * get_idf(token) * tf_token / (k1 * (1 - b * (1 - len(corpus[paper_id][where]) / average_doc_length)) + tf_token)
                pass

    if method == 'ltc-lnc':
        for paper_id in doc_scores:
            doc_scores[paper_id] /= math.sqrt(norm_factors[paper_id])
    return doc_scores
```

```python
from typing import List


def calculate_precision(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the precision of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The precision of the predicted results
    """

    # TODO: Calculate precision here
    valid_count, total_count = 0, 0
    for actual_list, predicted_list in zip(actual, predicted):
        actual_set = set(actual_list)
        predicted_set = set(predicted_list)

        total_count += len(predicted_set)
        valid_count += len(actual_set.intersection(predicted_set))

    return valid_count / total_count if total_count else 0
```

```python
def calculate_recall(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the recall of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The recall of the predicted results
    """

    # TODO: Calculate recall here
    valid_count, total_count = 0, 0
    for actual_list, predicted_list in zip(actual, predicted):
        actual_set = set(actual_list)
        predicted_set = set(predicted_list)

        total_count += len(actual_set)
        valid_count += len(actual_set.intersection(predicted_set))

    return valid_count / total_count if total_count else 0
```

```python
def calculate_F1(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the F1 score of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The F1 score of the predicted results    
    """

    # TODO: Calculate F1 here
    precision = calculate_precision(actual, predicted)
    recall = calculate_recall(actual, predicted)

    return 2 * precision * recall / (precision + recall) if precision + recall else 0
```

```python
def calculate_AP(actual: List[str], predicted: List[str]) -> float:
    precision, valid_count = 0, 0
    for index, value in enumerate(predicted):
        if value in actual:
            valid_count += 1
            precision += valid_count / (index + 1)

    return precision / valid_count if valid_count else 0

def calculate_MAP(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Mean Average Precision of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The Mean Average Precision of the predicted results
    """
    map = 0.0

    # TODO: Calculate MAP here
    for actual_list, predicted_list in zip(actual, predicted):
        map += calculate_AP(actual_list, predicted_list) / len(actual)

    return map
```

```python
import numpy as np

def calculate_NDCG(actual: List[List[str]], predicted: List[List[str]], k=10) -> float:
    """
    Calculates the Normalized Discounted Cumulative Gain (NDCG) of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The NDCG of the predicted results
    """
    ndcg = 0.0

    # TODO: Calculate NDCG here
    for actual_list, predicted_list in zip(actual, predicted):
        if k:
            actual_list = actual_list[:k]
            predicted_list = predicted_list[:k]

        rel = np.array([item in actual_list for item in predicted_list])
        dcg = np.sum((2 ** rel - 1) / np.log2(np.arange(2, rel.size + 2)))
        idcg = np.sum((2 ** np.sort(rel)[::-1] - 1) / np.log2(np.arange(2, rel.size + 2)))
        
        if idcg:
            ndcg += dcg / idcg

    return ndcg / len(actual)
```

```python
def calculate_MRR(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Mean Reciprocal Rank of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The MRR of the predicted results
    """

    # TODO: Calculate MRR here
    rank_sum = 0.0
    for actual_list, predicted_list in zip(actual, predicted):
        query_rank = 0.0
        relevant_docs = set(actual_list)

        for index, value in enumerate(predicted_list):
            if value in relevant_docs:
                query_rank = 1 / (index + 1)
                break

        rank_sum += query_rank

    return rank_sum / len(actual)
```
